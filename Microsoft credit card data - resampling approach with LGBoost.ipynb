```python
#Splitting dataset between X and Y. Creating a second dataset that contains all the columns with dtypes int64 and float64. These 
#columns have to be standardized
X = data_dummy_var.drop('Label',axis = 1)
Y = data_dummy_var['Label']
num_var= (X.dtypes[X.dtypes=='int64']).index | (X.dtypes[X.dtypes=='float64']).index
```

```python
scaled_df = pd.DataFrame(scaled_df, columns=['accountAge', 'accountCountry - Class000_LogOdds',
       'accountPostalCode - Class000_LogOdds',
       'accountState - Class000_LogOdds', 'browserLanguage - Class000_LogOdds',
       'browserType', 'cardNumberInputMethod', 'digitalItemCount',
       'ipCountry - Class000_LogOdds', 'ipPostalCode - Class000_LogOdds',
       'ipState - Class000_LogOdds', 'localHour - Class000_LogOdds',
       'numPaymentRejects1Day', 'paymentBillingCountryCode - Class000_LogOdds',
       'paymentBillingPostalCode - Class000_LogOdds',
       'paymentBillingState - Class000_LogOdds',
       'paymentInstrumentAgeInAccount', 'physicalItemCount',
       'purchaseProductType', 'responseCode',
       'shippingCountry - Class000_LogOdds',
       'shippingPostalCode - Class000_LogOdds',
       'shippingState - Class000_LogOdds', 'sumPurchaseAmount1Day',
       'sumPurchaseAmount30Day', 'sumPurchaseCount1Day',
       'sumPurchaseCount30Day', 'transactionAmount', 'transactionAmountUSD',
       'transactionCurrencyCode - Class000_LogOdds', 'transactionDate',
       'transactionDeviceType', 'transactionMethod', 'transactionTime'])
```

```python
scaled_df.head(10)
```

```python
X = X.drop(X[num_var],axis = 1) #Dropping all the initial non standardized columns with types = float64 and int64
```

```python
X = pd.concat([X,scaled_df],axis =1) #Merging the X with the scaled_df containing the standardized columns 
```

```python
cols = X.select_dtypes(exclude=['float','uint8']).columns #Converting the columns with object types to float

X[cols] = X[cols].apply(pd.to_numeric, downcast='float', errors='coerce')
```

```python
standardized_data = pd.concat([X,Y],axis =1)
#Running a reampling approach on this data and see where it goes 
from sklearn.utils import resample
data_nonfraud = standardized_data[standardized_data.Label==0]
data_fraud = standardized_data[standardized_data.Label==1]

data_fraud_undersampled = resample(standardized_data[standardized_data.Label==1], 
                                 replace=False,    # sample without replacement
                                 n_samples=797,     # represent around 61% of fraudulent cases
                                 random_state=75) # reproducible results

# Undersample majority class
data_nonfraud_undersampled = resample(standardized_data[standardized_data.Label==0], 
                                 replace=False,    # sample without replacement
                                 n_samples=797,     # to match minority class
                                 random_state=70) # reproducible results
 
# Combine minority class with downsampled majority class
data_undersampled = pd.concat([data_nonfraud_undersampled, data_fraud_undersampled])
 
# Display new class counts
print(data_undersampled.Label.value_counts())


#Creating new dataframe which does not contain the data that is present in creditcard_undersampled dataset
data_original_leftover = standardized_data.loc[~standardized_data.set_index(list(standardized_data.columns)).index.isin(data_undersampled.set_index(list(data_undersampled.columns)).index)]
#Checking shape of all datasets
print ('Dimension of data_original_leftover:'+str(data_original_leftover.shape))
print('Dimension of data_undersampled:',str(data_undersampled.shape))
print('Dimension of standardized_data:',str(standardized_data.shape))
```

```python
#Dataset should be ready to be fed to algorithm
from sklearn.model_selection import train_test_split 
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score,recall_score,f1_score,roc_curve, auc
X1 = data_undersampled.drop('Label',axis = 1)
Y1 = data_undersampled['Label']
X1_train, X1_test, Y1_train,Y1_test=train_test_split(X1,Y1,test_size=0.3, random_state=55)

#Using LGBM to train on dataset
import lightgbm as lgb
d_train = lgb.Dataset(X1_train, label=Y1_train)
params = {}
params['learning_rate'] = 0.003
params['boosting_type'] = 'gbdt'
params['objective'] = 'binary'
params['metric'] = 'binary_logloss'
params['sub_feature'] = 0.5
params['num_leaves'] = 10
params['min_data'] = 50
params['max_depth'] = 10

clf = lgb.train(params, d_train, 100) #Train the model with 100 iterations 

#Predict on dataset
pred_y1 = clf.predict(X1_test)
#convert into binary values
#Output will be a list of probabilities. I converted probabilities to binary prediction keeping threshold=0.5
for i in range(len(pred_y1)):
    if pred_y1[i]>=.5:       # setting threshold to .5
        (pred_y1[i])=1
    else:  
        (pred_y1[i])=0
print (pred_y1)
##Accuracy and Recall scores. 
print ('Accuracy score of LGBM classifier on test set:{:.3f}'.format(accuracy_score(Y1_test,(pred_y1))))
print ('Recall score of LGBM classifier on test set:{:.3f}'.format(recall_score(Y1_test,pred_y1)))
print ('F1 score of LGBM classifier on test set:{:.3f}'.format(f1_score(Y1_test,pred_y1)))
print()
cm_lgbm = confusion_matrix(Y1_test,pred_y1)
print('Confusion matrix with LGBM classifier with on undersampled test set:\n%s' % cm_lgbm)


```

```python
X2 = data_original_leftover.drop('Label',axis = 1)
Y2 = data_original_leftover['Label']

#Predicting on left over dataset 
pred_y1_leftover = clf.predict(X2)
#convert into binary values
#Output will be a list of probabilities. I converted probabilities to binary prediction keeping threshold=0.5
for i in range(len(pred_y1_leftover)):
    if pred_y1_leftover[i]>=.5:       # setting threshold to .5
        (pred_y1_leftover[i])=1
    else:  
        (pred_y1_leftover[i])=0


##Accuracy and Recall scores. 
print ('Accuracy score of LGBM classifier on test set:{:.3f}'.format(accuracy_score(Y2,pred_y1_leftover)))
print ('Recall score of LGBM classifier on test set:{:.3f}'.format(recall_score(Y2,pred_y1_leftover)))
print ('F1 score of LGBM classifier on test set:{:.3f}'.format(f1_score(Y2,pred_y1_leftover)))
print()
cm2_lgbm = confusion_matrix(Y2,pred_y1_leftover)
print('Confusion matrix with LGBM classifier with on undersampled test set:\n%s' % cm2_lgbm)
```
